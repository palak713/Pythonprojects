TextBlob is a python library for processsing the text data

```python
from textblob import TextBlob
words=["colleg","schol"]
corrected=[]
for i in words:
  corrected.append(TextBlob(i))
print("Wrong Word",words)
print("Corrected word are")
for i in corrected:
  print(i.correct())  
```
