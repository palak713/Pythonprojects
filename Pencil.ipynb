```python
gray = cv2.imread('/content/drive/MyDrive/new/1875ccd9356a4cf8e74c484846c749e5.jpg',0)
plt.imshow(gray)
plt.show()
```

```python
inverted_image=255-gray
burred=cv2.GaussianBlur(inverted_image,(21,21),0)
inverted_blurred=255-burred
pencil=cv2.divide(gray,inverted_blurred,scale=257)
plt.imshow(pencil)
plt.show()
```
